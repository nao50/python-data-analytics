```
xmin, xmax = -np.pi, np.pi
x = np.arange(xmin, xmax, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.hlines([-1, 1], xmin, xmax, linestyles="dashed")  # y=-1, 1
plt.title(r"$\sin(x)$ and $\cos(x)$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)
```

```
# sin
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title(r"$\sin x$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)

# cos
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, "r")
plt.title(r"$\cos x$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)

plt.tight_layout()
```

```
import numpy as np
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States"
df = pd.read_html(url, header=0)[0]
df
```

```
df['Location']
```

```
locations = df['Location'].str.extract(u'(\D+) (\d+\d+[NS]) (\d+\d+[WE]).*')
locations.columns = ['State', 'lat', 'lon']
locations
```

```
locations['lat'] = locations['lat'].str.replace(u'', '.')
locations['lon'] = locations['lon'].str.replace(u'', '.')

locations.loc[locations['lat'].str.endswith('S'), 'lat'] = '-' + locations['lat']
locations.loc[locations['lon'].str.endswith('W'), 'lon'] = '-' + locations['lon']

locations
```

```
locations['lat'] = locations['lat'].str.slice_replace(start=-2)
locations['lon'] = locations['lon'].str.slice_replace(start=-2)
locations[['lat', 'lon']] = locations[['lat', 'lon']].astype(float)

locations
```

```
locations.dtypes
```

```
df = pd.concat([df, locations], axis=1)
df
```

```
m = folium.Map(location=[55.0, -108.0], zoom_start=3.0)

for i, r in df.iterrows():
  folium.Marker(location=[r['lat'], r['lon']]).add_to(m)
    
m
```
